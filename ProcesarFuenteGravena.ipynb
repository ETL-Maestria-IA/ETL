```python
import requests
import json
import pandas as pd
import time  # Importar la librer√≠a para medir el tiempo

# Iniciar el temporizador
start_time = time.time()

# Convertir el dataframe a formato JSON (solo los registros)
data_json = df_soil_data_grevena.to_dict(orient='records')

# URL de la API con la clave de API (cambia la IP y la clave seg√∫n corresponda)
url = "http://3.123.234.166/api.php?api_key=Colombia25*"

# Enviar los datos a la API mediante POST
response = requests.post(url, json=data_json, headers={'Content-Type': 'application/json'})

# Finalizar el temporizador
end_time = time.time()

# Calcular el tiempo en minutos
elapsed_time = (end_time - start_time) / 60

# Verificar el estado de la respuesta y el contenido
print("Status Code:", response.status_code)
print("Response Text:", response.text)

# Ver la respuesta de la API
try:
    print(response.json())
except json.JSONDecodeError:
    print("Error al decodificar la respuesta JSON.")

# Mostrar el tiempo de carga en minutos
print(f"Tiempo de carga: {elapsed_time:.2f} minutos")
```
