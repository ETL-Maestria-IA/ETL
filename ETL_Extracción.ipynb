```python
from arcgis.gis import GIS
import pandas as pd

gis = GIS()
url = "https://services9.arcgis.com/Np9UOdtKcSTyWrE5/arcgis/rest/services/Propiedades_de_Suelos_Cuenca_de_Huanquisco/FeatureServer/0"

# Cargar los datos desde ArcGIS
feature_layer = gis.content.get(url)

# Obtener los datos como un DataFrame para el an√°lisis
df_soil_data_huanquisco = pd.DataFrame.spatial.from_layer(feature_layer)
```
